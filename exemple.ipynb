# Exemple d'utilisation d'OchunGRN
### I-Création de Graphe
La création de graphe aléatoire scale-free se fait avec la fonction `plotGraph`

```python
import numpy as np
import ochunGRN as oGRN
import random as rd
```

```python
GENESNB = 7
test = oGRN.randomGrn(GENESNB, 0.2, 0.4)
oGRN.plotGraph(test, actInhPlotBool=True)
```

```python

Coeff = oGRN.getCoefficient(GENESNB)
```

```python
oGRN.simulationODEs(test, ["massAction", "indirect", "Hill"], (0,100), Coeff)
oGRN.plotSim(test, ODEs="massAction", saveName=f"testmassAction.png")
```

```python
oGRN.plotSim(test, ODEs="indirect", saveName=f"testindirect.png")
```

```python
oGRN.plotSim(test, ODEs="Hill", saveName=f"testHill.png")
```

```python
oGRN.simulationODEs(test, ["Hill"], (0,100), Coeff)
oGRN.plotSim(test, saveName="test.png")
```

```python
FFLmotif = oGRN.randomGrn(100,0.1,0.02)
DictFFL=FFLmotif["subGraph"]
motifs = {v:0 for v in DictFFL.values()}
N=len(DictFFL)
print(N)
for i in DictFFL.keys():
    motifs[DictFFL[i]]+=1
    if DictFFL[i]=='FFL':
        print(i)
print(sorted(motifs.items(), key=lambda item: item[1]))
print(motifs['FFL']/N)
```
