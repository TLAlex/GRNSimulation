# Exemple d'utilisation d'OchunGRN
### I-Création de Graphe
La création de graphe aléatoire scale-free se fait avec la fonction `plotGraph`

```python
import numpy as np
import ochunGRN as oGRN
import random as rd
```

```python
GENESNB = 8
test = oGRN.randomGrn(GENESNB, 0.2, 0.1)
oGRN.plotGraph(test)
```

```python

Coeff = oGRN.getCoefficient(GENESNB)
```

```python
oGRN.simulationODEs(test, "massAction", (0,100), Coeff)
oGRN.plotSim(test, saveName=f"testmassAction.png")
```

```python
oGRN.simulationODEs(test, "indirect", (0,100), Coeff)
oGRN.plotSim(test, ODEs="indirect", saveName=f"testindirect.png")
```

```python
oGRN.plotProt(test, saveName="testindirectProt.png")
```

```python
oGRN.simulationODEs(test, "Hill", (0,100), Coeff)
oGRN.plotSim(test, ODEs="Hill", saveName=f"testHill.png")
```

```python
oGRN.simulationODEs(test, ["Hill"], (0,100), Coeff)
oGRN.plotSim(test, saveName="test.png")
```
