# Exemple d'utilisation d'OchunGRN
### I-Création de Graphe
La création de graphe aléatoire scale-free se fait avec la fonction `plotGraph`

```python
import numpy as np
import ochunGRN as oGRN
import random as rd
GENESNB = 5
test = oGRN.randomGrn(GENESNB, 0.2, 0.4)
oGRN.plotGraph(test, actInhPlotBool=True)
```

```python

Coeff = oGRN.getCoefficient(GENESNB)
```

```python
oGRN.simulationODEs(test, ["massAction", "indirect", "Hill"], (0,100), Coeff)
oGRN.plotSim(test, "test.png")
```

```python
oGRN.simulationODEs(test, ["indirect"], (0,100), Coeff)
oGRN.plotSim(test, "test.png")
```
